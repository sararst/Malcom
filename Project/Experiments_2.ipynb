```python
import networkx as nx
import matplotlib.pyplot as plt

def create_star_with_clusters():
    G = nx.Graph()
    
    # Central node
    central_node = 0
    G.add_node(central_node)
    
    # Create clusters at the end of each branch
    cluster_size = 5
    num_clusters = 5
    current_node = 1
    
    for i in range(num_clusters):
        # Create a cluster
        cluster_nodes = list(range(current_node, current_node + cluster_size))
        G.add_nodes_from(cluster_nodes)
        
        # Connect the cluster nodes to the central node
        for node in cluster_nodes:
            G.add_edge(central_node, node)
        
        # Create a complete subgraph (cluster)
        for j in range(cluster_size):
            for k in range(j + 1, cluster_size):
                G.add_edge(cluster_nodes[j], cluster_nodes[k])
        
        current_node += cluster_size
    
    return G

# Create the specific topology
G = create_star_with_clusters()

# Visualize the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=False, node_color='skyblue', node_size=500, edge_color='black')
plt.title("Star Topology with Clusters")
plt.show()
```
